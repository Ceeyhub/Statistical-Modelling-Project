Build a regression model.

```python
import pandas as pd
import sqlite3
import statsmodels.api as sm
```

```python
df = pd.read_csv('../data/merged_data.csv')
df
```

```python
df.dtypes
```

Provide model output and an interpretation of the results. 

```python
from sklearn.preprocessing import LabelEncoder
cols = ['free_bikes', 'review_count']
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
X = df['free_bikes']
y = df['review_count']
```

```python
X = sm.add_constant(X)
model = sm.OLS(y,X).fit()
model.summary()
```

From the P value of 0.000 and R-squared of 0.190 - the model is not a good fit
